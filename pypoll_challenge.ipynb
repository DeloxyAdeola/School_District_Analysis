```python

import pandas as pd
import os
```

```python
#Import the data from the CSV file
student_data = os.path.join('Resource','new_full_student_data.csv')
student_df = pd.read_csv(student_data)
```

```python
#Confirm that Pandas correctly imported the data by using
student_df.head()
```

```python
#returning the missing value with a boolean data type 
student_df.isnull()
```

```python
#Deliverable 2
```

```python
#check for rows that have NaN (or missing) values.
student_df.isnull().sum()
```

```python
#remove those rows that have NAN values
student_df = student_df.dropna()
student_df.isna().sum()
```

```python
#check for duplicate rows
student_df.duplicated().sum()
```

```python
#remove duplicated rows
student_df = student_df.drop_duplicates()
student_df.duplicated().sum()
```

```python
#Check the data types of the columns 
student_df.dtypes
```

```python
#In the grade column, remove the "th" suffix from every value
student_df["grade"] = student_df["grade"].str.replace("th", "")

student_df
```

```python
#Change the "grade" column to the int type,
student_df['grade'] = student_df['grade'].astype(int)
student_df.dtypes
```

```python
#Deliverable 3
```

```python
#summary statistics for the student DataFrame
student_df.describe()
```

```python
#Display the mean math score by using the mean function
student_df["math_score"].mean()
```

```python
#minimum reading score
min_reading_score = student_df["reading_score"].min()
min_reading_score
```

```python
#Deliverable 4
```

```python
#Display the grade column by using loc, 
student_df.loc[:,"grade"]
```

```python
#Display the first three rows of Columns 3, 4, and 5 by using iloc, 
student_df.iloc[:3,3:6]
```

```python
#Select the rows for Grade 9, and display their summary statistics 
grade_9_df = student_df.loc[student_df["grade"]== 9]
grade_9_df.describe()
```

```python
#Store the row with the minimum overall reading score. 
min_reading_row = student_df.loc[student_df["reading_score"]== min_reading_score]
min_reading_row
```

```python
#Select all the reading scores from the 10th graders at Dixon High School 
reading_score = student_df.loc[(student_df["grade"]== 10) & (student_df["school_name"]=="Dixon High School"),["school_name","reading_score"]]
reading_score
```

```python
#Find the mean reading score for all the students in Grades 11 and 12 combined by using conditional statements and loc or iloc.
reading_score = student_df["reading_score"].loc[(student_df["grade"]==11)|(student_df["grade"]==12)].mean()
            
reading_score
```

```python
#Deliverable 5
```

```python
#grouping 
student_group_df = student_df.groupby("school_type")
student_group_df
```

```python
#Display the average budget for each school type 
student_group_df = student_df[['school_type','school_budget']].groupby("school_type")

student_group_df.mean()
```

```python
#Find the total number of students at each school, and sort those numbers from largest to smallest by using the groupby, count, and sort_values functions,

student_school_name = student_df[["school_name","student_id"]].groupby(by='school_name').count()
student_count_df =pd.DataFrame(student_school_name)
student_count_df =student_count_df.rename(columns={"student_id": "student_count"})
student_count_df = student_count_df.sort_values(by=['student_count'], ascending = False)
student_count_df
```

```python
#Find the average math score by grade for each school type by using the groupby and mean functions
ave_math_grade_df = student_df.groupby(["school_type", "grade"])
ave_math_grade_df.mean()
```
